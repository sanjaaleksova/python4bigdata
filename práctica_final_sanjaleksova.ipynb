```python
# Importamos el dataset de los datos de los pokemon
df = pd.read_csv('dataset_final.csv')

#df.head()
df.tail()
```

```python
'''
1. ¿Cuántos pokémon hay en el dataset?
'''
number_pokemons = len(df.Name.unique())
print("Hay un total de ",number_pokemons, " pokémon únicos.")
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''

poison_1 = df[df['Type 1'] == 'Poison']['Name'].count()

print("Hay ", poison_1, " pokémon del tipo 'Poison'.")
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
unique_poison = len(df['Type 1'].unique())
print("Hay ", unique_poison, "pokémon únicos del tipo 'Poison'." )
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
pokemon_types = df.groupby("Type 1")[['Name']].count()
print(pokemon_types)

'''
2.1 ¿cual es el tipo de pokémon con más pokémon? (Type 1)
'''
print()
most_type1 = pokemon_types.sort_values(by='Name', ascending=False).head(1)
print(most_type1)
```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''

df.sort_values(by='Speed', ascending=False).head(1)
```

```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100?
'''

defense_greaterthan_100 = df[df['Defense'] > 100]['Name'].count()
print("Hay ", defense_greaterthan_100, "pokémon con una defensa superior a 100.")

'''
4.1 ¿Cuántos pokemon tiene una velocidad superior a 100?
'''

speed_greaterthan_100 = df[df['Speed'] > 100]['Name'].count()
print("Hay ", speed_greaterthan_100, "pokémon con una velocidad superior a 100.")

```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100?
'''
both_greaterthan100 = df[(df['Speed'] > 100) & (df['Defense'] > 100)]['Name'].count()
print("Hay ",both_greaterthan100, "pokémon con una defensa superior a 100 y con velocidad superior a 100.")


```

```python
'''
6. Ordena el dataset por el tipo 1 y por el tipo 2
'''

df.sort_values(by=['Type 1', 'Type 2'])
```

```python
'''
7. Crea un nuevo dataset con los pokémon de tipo Water y Fire como primer tipo
'''

water_and_fire = df[df['Type 1'].isin(['Water', 'Fire'])]
water_and_fire
```

```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''

legendary = df[df['Legendary'] == True]
legendary.head(10)
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
avg_stats_l = legendary[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].describe()
avg_stats_l
```

```python
non_legendary = df[df['Legendary'] == False]

avg_stats_nl = non_legendary[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].describe()
avg_stats_nl

```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios

ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
Legendary       99      90       89       91       94     90
No Legendary    80      95       75       12       43     87

'''

mean_l = avg_stats_l.loc[['mean']].astype('int32')
mean_nl = avg_stats_nl.loc[['mean']].astype('int32')

avg = pd.concat([mean_l, mean_nl])
avg
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

total = reduce(lambda a, b: a+b, map(lambda a: articulos[a], compra), 0)
print(f'El precio total de la compra: {round(total,2)} euros.\n')

# B. Una función que calcule el precio total de la compra con un descuento del 10%
discount10 = reduce(lambda a, b: a + b*0.9, map(lambda a: articulos[a], compra),0)
print(f'El precio total con un 10% de descuento aplicado es de {round(discount10, 2)} euros.\n')

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

discount10ifover100 = reduce(lambda a, b: a+b if sum(map(lambda a: articulos[a], compra)) <= 100 else a + b*0.9, map(lambda a: articulos[a], compra),0)
print(f'Se aplica 10% de descuento para compras superiores a 100 euros.\nTotal de la compra: {round(discount10ifover100,2)} euros.\n')
# D. Una función que calcule el precio total aplicando el IVA (21%)
totaltaxincluded = reduce(lambda a, b: a+b*1.21, map(lambda x: articulos[x], compra), 0)
print(f'El precio total con IVA aplicado es de {round(totaltaxincluded,2)} euros.\n')

# E.lista los artículos cuyo precio es superior a 20€

itemsover20 = list(filter(lambda a: articulos[a] > 20, articulos))
print(itemsover20)
```
